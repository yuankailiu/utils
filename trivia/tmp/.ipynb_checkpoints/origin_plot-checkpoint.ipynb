```python
! pwd
import numpy as np
import matplotlib.pyplot as plt
```

```python
origin_file = '/home/ykliu/scratch/stasis/Z-2004_origin.log'

with open(origin_file, 'r') as f:
    lines = f.readlines()
    lats = []
    lons = []
    xs   = []
    ys   = []
    for line in lines:
        tmp = line.split()
        lats.append(tmp[1])
        lons.append(tmp[2])
        xs.append(tmp[6])
        ys.append(tmp[7])

lats = np.array(lats, dtype=float)
lons = np.array(lons, dtype=float)
xs   = np.array(xs, dtype=float)
ys   = np.array(ys, dtype=float)
```

```python
fontz = 16

plt.figure(figsize=[6,6])
plt.axvline(x = -120.88)
plt.axvline(x = -119.88)
plt.axhline(y =  35.32)
plt.axhline(y =  36.32)

plt.scatter(lons, lats, c='b', s=10)

plt.xlabel('Longitude', fontsize=fontz)
plt.ylabel('Latitude',  fontsize=fontz)
plt.tick_params(axis='both', which='major', labelsize=fontz)
plt.title('Origin locations', fontsize=fontz+6)
plt.show()
```

```python
x_mid = (np.min(xs)+np.max(xs))/2
y_mid = (np.min(ys)+np.max(ys))/2

plt.figure(figsize=[6,6])
plt.axvline(x = x_mid)
plt.axhline(y = y_mid)

plt.scatter(xs, ys, c='k', s=10)

plt.xlabel('X', fontsize=fontz)
plt.ylabel('Y',  fontsize=fontz)
plt.tick_params(axis='both', which='major', labelsize=fontz)
plt.title('Origin locations (x,y)', fontsize=fontz+6)
plt.show()

print('X min = ', np.min(xs))
print('X max = ', np.max(xs))
print('Y min = ', np.min(ys))
print('Y max = ', np.max(ys))
print('X mid = ', x_mid)
print('Y mid = ', y_mid)
```
