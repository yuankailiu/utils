```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assume residuals of time series is a zero-mean Gaussian distribution
mu, sigma = 0, 1 # mean and standard deviation
r = np.random.normal(mu, sigma, (10000, 500))   # (N_pixels, N_acquisitions)

# Calculate the RMS of residual for each acquisition
rms = np.zeros(r.shape[1])
for i in range(r.shape[1]):
    rms[i] = (np.sum(r[:,i]**2)/r.shape[0])**0.5


med_r = np.median(r)                      # median of residual
med_rms = np.median(rms)                  # median of RMS residual

# Calculate the Median Absolute Deviation
MAD_r = np.median(np.abs(r-med_r))        # MAD of the residual
MAD_rms0 = np.median(np.abs(rms-mu))      # MAD of RMS residual with respect to zero mean of the residual
MAD_rms1 = np.median(np.abs(rms-med_rms)) # MAD of RMS residual with respect to median of the RMS residual


print('Median of residual = ', med_r)
print('Median of RMS = ', med_rms)

print('MAD of residual = ', MAD_r)
print('MAD of RMS residual (centered at 0) = ', MAD_rms0)
print('MAD of RMS residual (centered at median of RMS) = ', MAD_rms1)
```

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=[10,8])
bins = np.linspace(-5,5,100)
plt.hist(r.flatten(), bins)
plt.xlabel('Residual [mm]')
plt.ylabel('Samples')
plt.show()
```
